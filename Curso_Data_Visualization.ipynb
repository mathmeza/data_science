```python
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))

ax.bar(indice, temperatura_por_dia_da_semana)
ax.set_title("Temperatura por dia da semana", fontsize=15, pad=10)
ax.set_xlabel("Dia da Semana", fontsize=15)
ax.set_ylabel("Temperatura Média", fontsize=15);
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))
cores = ['black', 'red', 'g', 'b', 'yellow', 'orange', 'magenta']

ax.bar(indice, temperatura_por_dia_da_semana, color=cores)
ax.set_title("Temperatura por dia da semana", fontsize=15, pad=10)
ax.set_xlabel("Dia da Semana", fontsize=15)
ax.set_ylabel("Temperatura Média", fontsize=15)
ax.set_xticks(indice)
ax.set_xticklabels(nome_dias);
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))
cores = ['grey', 'red', 'g', 'b', 'yellow', 'orange', 'magenta']

ax.bar(indice, temperatura_por_dia_da_semana, color=cores, edgecolor='k')
ax.set_title("Temperatura por dia da semana", fontsize=15, pad=10)
ax.set_xlabel("Dia da Semana", fontsize=15)
ax.set_ylabel("Temperatura Média", fontsize=15)
ax.set_xticks(indice)
ax.set_xticklabels(nome_dias);
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

ax.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index)
ax.set_title('Temperatura por dia da semana', fontsize=15, pad=10);
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

ax.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
       autopct='%.1f%%')
ax.set_title('Temperatura por dia da semana', fontsize=15, pad=10);
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

explodir = [0.1,0,0,0,0,0.1,0.1]

ax.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
       autopct='%.1f%%', explode=explodir)
ax.set_title('Temperatura por dia da semana', fontsize=15, pad=10);
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

explodir = [0.1,0,0,0,0,0.1,0.1]

ax.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index,
       autopct='%.1f%%', explode=explodir, shadow=True)
ax.set_title('Temperatura por dia da semana', fontsize=15, pad=10);
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0,0,1,1])

ax.scatter(df['comprimento_sépala'], df['largura_sépala'])
ax.set_title("Gráfico de Dispersão", fontsize=25, pad=15)
ax.set_xlabel("Comprimeto da Sépala", fontsize=15)
ax.set_ylabel("Largura da Sépala", fontsize=15);
```

```python
fig = plt.figure(figsize=(15,7))
ax = fig.add_axes([0,0,1,1])

ax.scatter(df['comprimento_sépala'], df['largura_sépala'])
ax.set_title("Gráfico de Dispersão", fontsize=25, pad=15)
ax.set_xlabel("Comprimeto da Sépala", fontsize=15)
ax.set_ylabel("Largura da Sépala", fontsize=15)
ax.tick_params(labelsize=15);
```

```python
fig = plt.figure(figsize=(15,7))
ax = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa':'r', 'Iris-versicolor':'b', 'Iris-virginica': 'g'}

for especie in df['espécie'].unique():
  tmp = df[df['espécie']==especie]
  ax.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
             color=cores[especie])

ax.set_title("Gráfico de Dispersão", fontsize=25, pad=15)
ax.set_xlabel("Comprimeto da Sépala", fontsize=15)
ax.set_ylabel("Largura da Sépala", fontsize=15)
ax.tick_params(labelsize=15)
ax.legend(cores, fontsize=20);
```

# Customizando Histogramas

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

ax.hist(df['comprimento_pétala'], bins=20)
ax.set_title('Histograma', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)

ax.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu, sigma),
            xy=(4.5, 20), fontsize=20);
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

ax.hist(df['comprimento_pétala'], bins=20)
ax.set_title('Histograma', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)

ax.annotate('$mu = {0:.2f}$\n$sigma = {1:.2f}$'.format(mu, sigma),
            xy=(4.5, 20), fontsize=20);

ax.axvline(mu, color='k', linestyle='--')
ax.annotate('Média', xy=(mu-1.3, 28), fontsize=20);            
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

ax.hist(df['comprimento_pétala'], bins=20)
ax.set_title('Histograma', fontsize=15, pad=10)
ax.set_xlabel('Comprimento da pétala', fontsize=15)
ax.grid(True)

ax.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
            xy=(4.5, 20), fontsize=20);

ax.axvline(mu, color='k', linestyle='--')
ax.annotate('Média', xy=(mu-1.3, 28), fontsize=20)

ax.axvline(df['comprimento_pétala'].median(), color='g', ls='--')
ax.annotate('Mediana', xy=(df['comprimento_pétala'].median(), 31), fontsize=20);            
```

```python
# Criando Figura e Eixo
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

# Instanciando cálculo de média e desvio padrão nas variáveis
mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

# Configurando gráfico
ax.hist(df['comprimento_pétala'], bins=20)
ax.set_title('Histograma Comprimento Pétala', fontsize=15)
ax.set_xlabel('Comprimento da pétala', fontsize=13)
ax.grid(True)

# Adicionando linhas verticais e anotações
ax.axvline(mu, ls='--', color='k')
ax.annotate('Média', xy=(mu-1, 20), fontsize=15)

ax.axvline(df['comprimento_pétala'].median(), ls='--', color='g')
ax.annotate('Mediana', xy=(df['comprimento_pétala'].median(), 28), fontsize=15)

# Adicionar valor da Média e Desvio Padrão no gráfico.
ax.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma), xy=(5,20), fontsize=15);
```

```python
# Criando Figura e Eixo
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

df_iv = df[df['espécie']== 'Iris-versicolor']

# Instanciando cálculo de média e desvio padrão nas variáveis
mu, sigma = df_iv['comprimento_pétala'].mean(), df_iv['comprimento_pétala'].std()

# Configurando gráfico
ax.hist(df_iv['comprimento_pétala'], bins=20)
ax.set_title('Iris Versicolor', fontsize=15)
ax.set_xlabel('Comprimento da pétala', fontsize=13)
ax.grid(True)

# Adicionando linhas verticais e anotações
ax.axvline(mu, ls='--', color='k')
ax.annotate('Média', xy=(mu-0.5, 5.5), fontsize=15)

ax.axvline(df_iv['comprimento_pétala'].median(), ls='--', color='g')
ax.annotate('Mediana', xy=(df_iv['comprimento_pétala'].median()-0.7, 6.5), fontsize=15, color='g')

# Adicionar valor da Média e Desvio Padrão no gráfico.
ax.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma), xy=(4.6,6), fontsize=15)

fig.savefig('histograma_iv.png', bbox_inches='tight')
```
